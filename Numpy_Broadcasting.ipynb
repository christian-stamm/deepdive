## Exercise 2: Normalize features
Create `X_norm=(X-mean)/std` using broadcasting.

```python
import numpy as np
X=np.array([[1.,2.],[3.,6.],[5.,10.]])
mean=X.mean(axis=0)
std=X.std(axis=0)
```

```python
# YOUR ANSWER HERE
X_norm=(X-mean)/std
```

```python
assert 'X_norm' in globals();
np.testing.assert_allclose(X_norm,(X-mean)/std)
print('✅ Correct!')
```

## Exercise 3: Batch logits
Compute `logits = X*w` then sum features and add bias using broadcasting.

```python
import numpy as np
X=np.arange(12.).reshape(3,4)
w=np.array([1.,0.,-1.,2.])
b=0.5
```

```python
# YOUR ANSWER HERE
logits= X @ w.T + b
```

```python
assert 'logits' in globals();
exp=(X*w).sum(axis=1)+b
np.testing.assert_allclose(logits,exp)
print('✅ Correct!')
```

## Challenge
Implement softmax using broadcasting.

```python
import numpy as np
scores=np.array([[1.,2.,3.],[1.,1.,1.]])
```

```python
# create probs
exps = np.exp(scores - np.max(scores, axis=1, keepdims=True))
probs = exps / np.sum(exps, axis=1, keepdims=True)
```

```python
assert 'probs' in globals();
exp=np.exp(scores-scores.max(axis=1,keepdims=True));ans=exp/exp.sum(axis=1,keepdims=True)
np.testing.assert_allclose(probs,ans)
print('✅ Softmax correct!')
```
